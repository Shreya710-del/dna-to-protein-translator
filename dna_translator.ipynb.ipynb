```python
# Step 1: Store a DNA sequence
dna_sequence = "ATGGCCGAA"
# Step 2: Print it
print("DNA Sequence:", dna_sequence)

# Step 3: Convert DNA to RNA
rna_sequence = dna_sequence.replace("T", "U")

# Step 4: Print RNA sequence
print("RNA Sequence:", rna_sequence)
# Step 5: Codon to amino acid dictionary
codon_table = {
    "AUG": "Methionine", "GCC": "Alanine", "GAA": "Glutamic Acid",
    "UUU": "Phenylalanine", "UUC": "Phenylalanine", "UUA": "Leucine", "UUG": "Leucine",
    "UAA": "STOP", "UAG": "STOP", "UGA": "STOP"
    # You can add more codons later
}

# Step 6: Split RNA into codons (3 letters at a time)
codons = [rna_sequence[i:i+3] for i in range(0, len(rna_sequence), 3)]

# Step 7: Translate codons to protein
protein_sequence = []
for codon in codons:
    amino_acid = codon_table.get(codon, "Unknown")
    protein_sequence.append(amino_acid)

# Step 8: Print final protein sequence
print("Protein Sequence:", "-".join(protein_sequence))
import matplotlib.pyplot as plt

# Simple bar chart for codons
plt.figure(figsize=(8, 4))
plt.bar(codons, [1]*len(codons), tick_label=protein_sequence, color="skyblue")
plt.title("Codon to Amino Acid Mapping")
plt.xlabel("Codons")
plt.ylabel("Amino Acids")
plt.grid(True)
plt.show()

```
